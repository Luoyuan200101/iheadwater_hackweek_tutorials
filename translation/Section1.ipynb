### 关于开放科学

#### 1. 什么是开放科学

开放科学涉及向所有人提供科学方法，数据和结果。它可以分解为几个部分（Gezelter 2009），包括：

- 数据收集，处理和分析方法以及结果推导的透明度。  
- 公开可用的数据和关联的处理方法。  
- 结果的透明沟通。  
- 开放科学也经常得到合作的支持。  

可重复的科学是指任何人（包括他人和你未来的自己）都可以理解和复制分析的步骤，应用于相同甚至新的数据。  

共同提供来自开放科学工作流的开放可重现科学结果，使您能够轻松地与他人共享工作和协作，以及公开发布您的数据和工作流，以增加科学知识。

#### 2. 开放可重复科学的优势

科学中开放性和可重复性的好处包括：

- 科学过程的透明度，因为包括公众在内的任何人都可以访问数据，方法和结果。  
- 于他人复制和扩展您的工作，这进一步支持了同行评审和协作学习。   
- 随着时间的推移，您可以根据需要随时轻松理解和重新运行自己的分析。  

#### 3. 你如何让你的作品更加开放和可重复？

**使用科学编程处理数据**

科学编程允许您自动执行任务，从而有助于快速运行和复制您的工作流程。相比之下，基于图形用户界面 （GUI） 的工作流需要交互式手动步骤进行处理，这变得更加困难和耗时。如果您使用像 或 这样的开源编程语言，那么任何人都可以访问您的方法。但是，如果您使用需要许可证的工具，则没有资源购买该工具的人员将被排除在完全复制您的工作流之外。Python、R

**使用文件和目录的富有表现力的名称来组织您的工作**

富有表现力的文件和目录名称允许您快速找到所需的内容，并且还通过促进其他人了解您的文件和工作流来支持可重复性（例如，名称可以告诉其他人文件或目录包含的内容及其用途）。确保将相关文件组织到目录（即文件夹）中，以帮助您轻松分类和查找所需的内容（例如.raw数据，脚本，结果）。

**使用公平数据提高项目的可重复性**

确保项目中使用的数据符合FAIR原则（Wilkinson et al. 2016），以便它们可查找，可访问，可互操作和可重用，并且有关于如何访问它们及其包含的内容的文档。FAIR原则还超越了原始数据，适用于用于处理和创建新数据的工具和工作流程。FAIR原则通过支持数据和工作流程的重用和扩展来增强项目的可重复性，这有助于在科学界内进行更大的发现。

**保护您的原始数据**

不要修改（或覆盖）原始数据。将数据输出与输入分开，以便您可以根据需要轻松重新运行工作流。如果您将数据组织到将原始数据与结果分开的目录中，则可以轻松完成此操作，等等。

**使用版本控制和共享代码**

版本控制允许您管理和跟踪对文件的更改（甚至撤消它们！如果可以公开共享代码，请实现版本控制，然后在云上发布代码和工作流。有许多免费工具可以做到这一点，包括Git和GitHub。

**记录您的工作流程**

文档可能意味着许多不同的事情。它可以是基本的，就像在整个代码中包括（精心设计的和切中要害的）注释一样，以解释工作流的特定步骤。文档还可能意味着使用Jupyter Notebooks或RMarkdown文件等工具来包含Markdown格式的文本叙述，该文本叙述穿插了代码以提供工作流的高级解释。

文档还可以包括文档字符串，它提供Python函数的标准化文档，甚至是描述工作流，目录结构，数据，处理和输出的更大图景的README文件。

**设计可轻松重新创建的工作流程**

您可以通过以下方式设计可由其他人轻松重新创建和复制的工作流：

- 在代码文件的顶部列出运行工作流所需的所有包和依赖项（例如 Jupyter Notebook 或 R Markdown 文件）。
- 将代码组织到相关代码的各个部分或代码块中，并包括注释来解释代码。
- 使用 Docker 容器、conda 环境和带绑定器的交互式笔记本等工具为 Python 工作流创建可重用的环境。

### 开放可重复科学的最佳实践

#### 1. 开放可重复科学项目的项目组织和管理

**组织您的科学项目目录，使其更易于理解**

在处理数据项目时，通常需要在计算机上存储许多文件。这些文件可能包括：
- 原始数据文件
- 已处理的数据文件：您可能需要获取原始数据并以某种方式对其进行处理
- 代码和脚本
- 图形和表格等输出
- 与您的项目相关的写作  
如果您仔细考虑这些文件是如何存储在您的计算机上的，它将节省您的时间并使您的项目更加可用和可重复。以下是在整合项目时要考虑的一些最佳实践。

**目录和文件名的重要性**

当您在计算机上创建新目录和文件时，请考虑使用精心设计的命名约定，这样任何人都可以更轻松地查找内容并了解每个文件的功能或包含的内容。

使用以下文件和目录是一个好习惯：

- 人类可读：使用表达性名称，清楚地描述目录或文件包含的内容（例如代码、数据、输出、图形）。 
- 机器可读：避免使用奇怪的字符或空格。-您可以使用或分隔名称中的单词而不是空格，_以使它们易于阅读和解析。  
- 可排序：能够对文件进行排序以快速查看其中的内容并找到您需要的内容非常好。例如，您可以为相关目录或文件的列表（例如 、 等）创建一个命名约定01-max.jpg，02-terry.jpg这将产生可排序的文件。  

这些指南不仅可以帮助您组织目录和文件，还可以帮助您实现机器可读的名称，这些名称可以使用科学编程或其他形式的脚本轻松查询或解析。

在构建项目目录时使用良好的命名约定还可以帮助不熟悉您的项目的其他人快速了解您的目录和文件结构，从而支持可重复性。

#### 2. 开放可重复科学项目的最佳实践

**使用一致的计算机可读命名约定**

机器可读的文件名允许您的目录结构由代码快速操作和处理。

例如，您可能想要编写一个处理一组图像的脚本，并且您可能想要按日期对这些图像进行排序。如果每个图像的日期都包含在名称开头的文件名中，那么使用您的代码解析将变得更容易。以下文件可能难以解析，因为命名约定不标准。

* file.jpg
* file-two.jpg
* filethree.jpg
 
然而，这个文件列表更容易解析，因为日期包含在文件名中。

* 2020-image.jpg
* 2019-image.jpg
* 2018-image.jpg

有时简单地对文件进行编号就足以进行排序：

* 01-image.jpg
* 02-image.jpg
* 03-image.jpg

如果您的文件和目录遵循可识别的模式或规则，它将允许您更轻松地操作它们。这反过来将使您更容易自动化文件处理任务。

在项目中命名文件时要考虑的其他一些最佳实践：

避免文件和目录名称中的空格：在自动化工作流程时，文件名中的空格可能很困难。
使用破折号来分隔单词（slugs）：破折号或下划线可以让您更轻松地创建富有表现力的文件名。破折号或下划线在编码时也更容易解析。
考虑您是否可能需要对文件进行排序。如果你这样做了，你可能想给事物编号。

**命名文件时保持一致 - 使用小写**

lower命名文件和目录以使用和Upper大小写时可能很诱人。但是，如果您在操作系统（Mac、Linux 和 Windows）之间切换，大小写会给您带来编码问题。

举个例子，看看下面的文件名。

* my-file.txt
* My-File.txt

如果你想打开/阅读my-file.txt它会很容易调用：

pandas.read.csv("my-file.txt")

在 Python 中。此调用适用于所有操作系统。但是，这个电话：

pandas.read.csv("My-file.txt")

可能适用于某些机器（可能是 Windows），但在 Linux 或 MAC 上可能会失败。为了简单起见并避免大小写敏感问题，请对所有文件和目录名称使用小写命名约定。

**组织您的项目目录，以便轻松查找数据、代码和输出**

与其将一堆文件保存到一个目录中，不如考虑一种适合您项目的目录组织方法。

创建涵盖工作流程步骤的编号目录 - 例如：

/vegetation-health-project  
    /01-code-scripts  
    /02-raw-data  
    /03-processed-data  
    /04-graphics-outputs

每个文件夹之前的数字允许您以更易于解析的方式对目录进行排序。另请注意，每个目录都有一个富有表现力的（使用描述目录中内容的词语）名称。下一节将讨论富有表现力的命名。

使用单独的目录来存储数据、脚本、输出图形以及为项目编写的最终论文和博客文章，可以更轻松地找到项目的组件。

这对于您未来的自己特别有用，他们可能需要在六个月后回到项目中进行更新。它还使您正在与之合作的同事更容易快速找到东西。

没有一个完美的例子，因为每个项目可能需要不同的目录。最好的建议是选择适合您和您的团队的东西并坚持下去。最好保持一致。



**使用有意义（表达性）的文件和目录名称**

富有表现力的文件名是有意义的，因此描述了每个目录或文件是什么或包含什么。使用富有表现力的文件名可以更轻松地扫描项目目录并快速了解存储的位置以及文件包含或包含的内容。

表达性名称还支持机器可读性，因为表达性名称中的可识别模式可以被计算机用来识别和解析文件。

**使用 README 文件记录您的项目**

有很多方法可以记录项目；但是，项目顶层的自述文件是标准约定。当您开始使用 GitHub 时，您会注意到几乎所有设计良好的 github 存储库都包含自述文件。自述文件是一个文本文件，描述了用于处理项目中数据的数据/软件包和工具。自述文件还应描述文件和相关的命名约定。最后，自述文件可用于根据需要记录使用的任何缩写、单位等。

您还可以考虑其他文件，例如软件安装说明（如果需要）、引用信息以及如果项目是您希望其他人参与的项目，那么CONTRIBUTING文件可能是有序的。

**不要使用专有文件格式**

专有格式是需要特定工具（通常是特定许可证）才能打开的格式。示例包括 Excel (.xls) 或 Word (.doc)。随着新版本的出现，这些格式可能会随着时间而改变（例如：.xls升级到.xlsx.

在某些情况下，某些格式是特定于操作系统的（例如：大多数 Linux 用户不运行 Microsoft 工具）。

在为您的项目选择文件格式时，请考虑您将来是否拥有访问该文件格式的许可，以及其他人是否有权访问该许可。

如果可以，请坚持使用与操作系统和工具无关的格式，例如.csv和.txt. 文本文件不是专有的，因此可以在任何操作系统和任何具有正确打开工具的计算机上打开。这允许更多人访问您的文件，包括您未来可能没有打开这些文件的许可证的自己。

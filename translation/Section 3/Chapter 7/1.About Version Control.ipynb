# 第1课：版本控制

## Ⅰ. 什么是版本控制

版本控制系统保持着对代码和其他内容的修改记录。它还允许我们将变化恢复到以前的时间点。


## Ⅱ. 版本控制的类型

有许多版本控制的形式。有些没有那么好:

- 用新的日期或名称保存文件（我们都做过，但效率不高，容易失去对最新文件的追踪）。
- Google Docs的 "历史 "功能（对某些文件来说还不错，但范围有限）。

有些更好:

- 版本控制工具，如Git, Mercurial, 或Subversion。

## Ⅲ. 为什么版本控制很重要

版本控制促进了许多科学工作流程的两个重要方面:

1. 保存和审查或恢复到以前版本的能力。
2. 在一个项目上进行协作的能力。

这意味着你不必担心合作者（或你未来的自己）会改写一些重要的东西。它还允许在同一个文件上工作的两个人有效地结合想法和修改。

## Ⅳ. 版本控制系统是如何工作的

### 1. 简单的版本控制模式
版本控制系统跟踪一个或多个文件随着时间的推移所发生的变化。版本控制系统从一个文件的基本版本开始。然后，它们保存你所做的修改。

你可以把版本控制想象成一盘磁带：如果你把磁带倒回去，从基础文件开始，那么你可以回放每一个变化，最后得到你的最新版本。



版本控制系统在你向系统添加和提交文件时，会按顺序保存对文件的修改。

一旦你认为修改与文档本身是分开的，你就可以考虑在基础文档上 "回放 "不同的修改集。然后，你可以检索或恢复到文档的不同版本。

通过版本控制的协作，用户可以对同一个文档进行独立的修改。



同一文件的不同版本可以保存在一个版本控制系统中

如果用户的修改之间没有冲突（冲突是指两个用户以不同的方式修改同一个文档的相同部分），你可以在同一个基础文档上审查两套修改。如果有冲突，可以通过选择你想保留的修改来解决。



如果没有冲突（两个用户以不同的方式修改同一文件的同一部分），对同一基础文件的两组修改可以在版本控制系统中合并在一起。如果有冲突，可以通过选择你想保留的修改来解决。冲突解决后，所有由两个用户提交的其他修改都可以合并在一起。

版本控制系统是一种工具，为我们跟踪所有这些变化。一个文件的每个版本都可以在任何时候被查看和恢复。这样，如果你添加了你最终不喜欢的东西或删除了你需要的东西，你可以简单地回到以前的版本。

### 2. Git和GitHub - 一个分布式的版本控制模型

**Git** 采用分布式的版本控制模式。这意味着仓库可以有很多副本（或**GitHub**世界中的fork/branches）。在本地工作时，**git** 是你用来跟踪仓库变化的程序。

**GitHub.com** 是互联网上的一个位置（云网络服务器），作为你的仓库的一个远程位置。**GitHub** 为你的工作提供了一个备份，如果你的本地拷贝丢失了（比如你的电脑从码头上掉下来），你就可以检索到。**GitHub**还允许你分享你的工作，并与他人合作开展项目。



分布式版本控制系统的一个优点是有许多版本库的副本。因此，如果任何一台服务器或电脑死了，任何一个客户端的存储库都可以被复制并用来恢复数据

## Ⅴ. Git和GitHub如何支持版本控制

由于每个工具所提供的功能，你可以在同一个工作流程中同时使用git和GitHub，以。

- 在本地使用 git 跟踪你的代码的变化。
- 在不同版本之间同步代码（即自己的版本或他人的版本）。
- 测试对代码的修改，而不丢失原版。
- 如果需要的话，可以恢复到旧版本的代码。
- 在云端（GitHub.com）上备份你的文件。
- 在GitHub.com上分享你的文件并与他人协作。

在这本教科书中，你将了解更多关于git和GitHub在版本控制和协作方面的功能，以支持开放的可重复的科学。

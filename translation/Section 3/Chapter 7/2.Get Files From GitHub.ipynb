# 第2课：Copy (Fork) and Download (Clone) GitHub Repositories

## Ⅰ. 关于Git和GitHub

在这本教科书的前面，你了解到**git**是一种工具，通过一套你可以在终端执行的命令来跟踪文件的变化（这个过程称为版本控制）。你还了解到，**GitHub**允许你将文件存储在云端，以便从任何电脑上访问它们，并与他人分享。

你可以在工作流程中同时使用**git**和**GitHub**，用**git**在本地对文件进行修改，并在**GitHub.com**上存储和分享你的文件。为了协同工作，**git**和**GitHub**使用仓库（即文件的目录）来管理和存储文件。

将一组文件的副本存储在云端的GitHub仓库中是很理想的，因为：

1. 有一个备份。如果你的电脑出了问题，这些文件仍然可以在线使用。
2. 你可以很容易地与其他人分享这些文件。
3. 你甚至可以使用第三方工具（如Zenodo）创建一个数字对象标识符（DOI）来引用你的文件或要求别人引用你的文件。你还可以将这些DOI添加到你的简历或C.V.中，以促进你的工作。

## Ⅱ. 仓库的目录结构

从本质上讲，仓库是一个特定项目的目录，它被**git**和**GitHub**认定为仓库，因为它包含一个名为 .git 的子目录。

.git 子目录是自动创建的，如果是在 **GitHub.com** 上创建的，则由 **GitHub** 创建；如果是在电脑上先创建的仓库，则由 **git** 创建（即初始化为一个仓库）。这个 .git 子目录被这些工具用来管理和跟踪在这个目录上运行的各种任务（比如跟踪仓库中文件的变化）。因此，你永远不需要访问或修改 .git 子目录中的文件。

1. README.md文件  ； 这是一个Markdown文件，用来提供版本库的描述（即它的内容、目的等），以便其他人可以学习如何使用版本库中的文件。
2. .gitignore 文件 ： 这个文件可以用来列出你不希望git跟踪的文件（即通过版本控制监控）。你将在本章后面了解更多关于这两个有用的文件。

## Ⅲ. GitHub.com上仓库的URL

当一个仓库存储在**GitHub.com**上时，它会被分配一个唯一的URL（即GitHub.com网站上的链接），可以用来找到该仓库并访问其文件。虽然**GitHub.com**上的仓库可以是公开的，也可以是私有的，但对于免费的**GitHub**账户，默认是公开的。

无论哪种情况（公开或私有），**GitHub** 仓库的 URL 链接总是遵循相同的格式。

`https://github.com/username/repository-name`

用户名是该仓库创建者（即所有者）的用户名。这个用户名可以是个人，比如 `eastudent` （或者你的 **GitHub** 用户名！），也可以代表一个组织，比如 `earthlab-education`。

例如，在这本教科书中你将使用的资源库是由 `earthlab-education` 所有的，因此，它的URL看起来像这样。

`https://github.com/earthlab-education/repository-name`

## Ⅳ. 在GitHub.com上创建其他用户的文件副本（Fork a Repo）

使用**GitHub.com**，你可以复制一个由其他用户或组织拥有的**GitHub**仓库（也被称为 **repo**）,这项任务被称为fork仓库。这意味着**你不需要fork一个你已经拥有的仓库**。相反，其他用户可以fork你的版本库，如果他们想要一个副本来工作，而你的原始文件将不会被修改。

fork仓库的能力是使用 GitHub 仓库的一个好处，因为被fork的仓库是与原始仓库相连的。这意味着你（或其他用户）可以将你fork的文件从原版更新到你（或他们）的fork仓库。这也意味着你可以对原始版本库提出修改建议，并由该版本库的所有者审核。因此，fork允许你与他人协作，同时保护文件的原始版本。协作时，每个人都会用原始文件的副本工作。而且所有的修改都会在每个文件的历史中被跟踪，并可以在任何时候撤销。

你可以从你想复制的仓库的GitHub.com主页面fork一个现有的GitHub仓库。

fork仓库:

1. 导航到你想分叉的版本库页面 - 例如。
`https://github.com/earthlab-education/practice-git-skillz`

2. 在该页面上，你会看到右上角有一个按钮，上面写着Fork。该按钮旁边的数字告诉你该 repo 已经被fork了多少次。
3. 点击 "fork"按钮，当它询问你想在哪里fork该版本时，选择你的用户账户。
4. 一旦你fork了该 repo，你的账户中就会有一个副本。导航到你的 repo 页面。网址应该是这样的。

`https://github.com/your-user-name/practice-git-skillz`

要fork一个 repo，首先导航到你想fork的 repo。然后点击屏幕右上角的**fork**按钮。然后你就可以在你的账户中创建一个该 repo 的副本。
在这本教科书的后面，你将学习如何对原始版本库提出修改建议，从原始版本库接收更新到你的fork版本，并与他人合作。

## Ⅴ. 从GitHub.com复制文件到你的电脑（git clone）

要在本地使用 **GitHub** 仓库（包括fork仓库），你需要在你的电脑上创建该仓库的本地副本（这项任务被称为克隆仓库）。你可以克隆你自己的 **GitHub** 仓库，也可以克隆别人的仓库（比如你分叉到你的 **GitHub** 账户的仓库）。

无论哪种情况，克隆都可以创建一个 **GitHub** 仓库的本地副本，这样就可以在本地电脑上处理文件了。克隆仓库是在本地处理文件的好方法，同时在云端的 **GitHub.com** 上还有一份文件副本。按照下面的步骤，你将在**终端**使用 `git clone` 命令来克隆 **GitHub** 仓库。

### 1. 使用Bash来改变你所期望的工作目录

使用任何 **git** 命令的第一步是将当前工作目录改为你想要的目录。就 `git clone` 而言，当前工作目录需要是你想下载 **GitHub** 仓库的本地副本的地方。

在这本教科书中，你将克隆一个 repo 到你的电脑（或你工作的地方）上一个叫做 `earth-analytics` 的目录。这个 `earth-analytics` 目录应该位于你电脑的主目录中。



### 2. 从GitHub.com复制一个Github.com仓库的URL

要运行 `git clone` 命令，你需要你想克隆的仓库的 URL（即你拥有的仓库或你从其他用户的仓库创建的分叉）。

在该仓库的**GitHub.com**主页上，你可以点击`Clone or download`的绿色按钮，然后把提供的URL复制到框中，看起来就像:

`https://github.com/your-username/practice-git-skillz`

### 3. 在终端运行 Git 克隆命令

现在你有了你想复制到本地的版本库的URL，你可以用终端运行`git clone`命令，后面跟上你复制的URL:

***
git clone https://github.com/your-username/practice-git-skillz
***

现在你已经在你的 `earth-analytics` 目录下做了一个版本库的本地拷贝。你可以用**终端**的`ls`命令仔细检查这个目录是否存在。

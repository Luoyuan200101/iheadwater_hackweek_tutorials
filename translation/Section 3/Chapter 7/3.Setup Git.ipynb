# 第3课：如何在你的电脑上本地设置Git

## Ⅰ. 版本控制的Git和GitHub工作流程

之前，你学过如何`fork` **GitHub** 仓库以复制其他用户的仓库，也学过如何下载 **GitHub** 仓库的副本（即`clone`）到你的电脑。在这一页，你将学习如何在你的电脑上设置**git**

## Ⅱ. 在你的电脑上配置git用户名和电子邮件

第一次在电脑上使用 **git** 时，需要配置 **GitHub.com** 的用户名和电子邮件地址。这些信息将被用来记录谁对**git**中的文件进行了修改。使用你在**GitHub.com**上设置的相同的电子邮件地址和用户名是很重要的。

你可以在**终端**上输入**Github.com**的用户名来设置。

`$ git config --global user.name "username"`

接下来，你可以通过键入来设置你的**Github.com**账户的电子邮件。

`$ git config --global user.email "email@email.com"`

使用`--global`配置选项，你是在告诉**git**对你在电脑上工作的所有**git**仓库使用这些设置。注意，你只需要在你的电脑上配置一次这些设置。

你可以用以下命令检查你对`user.name`和`user.email`的配置。

`git config user.name`，它返回你之前设置的用户名

`git config user.email` 会返回您之前设置的电子邮件。

这些设置可以确保你对仓库所做的修改归于你的用户名和电子邮件。

## Ⅲ. 通过GitHub为Git设置身份验证

GitHub 要求通过认证才能对 repo 进行任何修改。有两种方法来设置GitHub的认证。

1. 使用个人访问标识，你可以在GitHub.com上设置并在本地使用该令牌进行认证。这涉及到在GitHub.com上创建一个标识，然后在本地bash中使用它作为 "密码"。
2. 使用SSH：这需要在本地进行更多的设置，但一旦设置好了，你就可以跳过每次对 repo 进行修改时的认证步骤。

在这一课中，你将了解这两个选项。如果你可以的话，我们确实推荐你使用SSH认证。然而，标识的方法也很好用。

### 1. 设置一个用于认证的GitHub标识

设置标识在短期内是比较快的，因为它需要较少的步骤。然而，从长远来看，基于标识的认证可能更难管理：

1. 每次需要验证GitHub账户时，都必须输入标识（除非你缓存/存储它）
2. 如果你丢失了你的标识，你必须刷新/停止一个新的标识。

### 2. 建立与GitHub的SSH连接（推荐）

SSH是Secure SHell的缩写，是一种从电脑上认证GitHub的替代方式。虽然SSH的设置涉及更多的步骤，但一旦设置好，你就不必再去验证与GitHub的连接。这是因为你的电脑上会有一个储存在本地的密钥，可以与储存在 GitHub 账户中的密钥进行验证。

设置 SSH 包括两个步骤：

1. 在你的电脑上创建密钥本身
2. 将密钥添加到你的 GitHub 账户。

下面是两个GitHub教程，指导你完成在电脑上设置SSH的所有步骤

**步骤1：在本地生成你的SSH密钥**

[Use this link to find step by step instructions for generating your SSH key](https://docs.github.com/en/authentication/connecting-to-github-with-ssh/generating-a-new-ssh-key-and-adding-it-to-the-ssh-agent)

请确保按照指示一直走到页面的最后，在那里你将你的SSH密钥添加到你的SSH代理。**注意**：在第4步左右，指示告诉你为你的SSH连接设置一个口令。如果你愿意，你可以这样做，但这是一个可选的步骤。SSH已经非常安全了，所以如果你决定不需要SSH口令，你可以在空的口令行上点击回车。

**步骤2：将你的SSH密钥添加到你的GitHub账户上**

生成密钥并将其添加到代理后，需要将密钥添加到 GitHub 账户中。[该链接提供了将 SSH 密钥添加到账户中的逐步说明](https://docs.github.com/en/authentication/connecting-to-github-with-ssh/adding-a-new-ssh-key-to-your-github-account)。各个操作系统的说明略有不同，所以请确保你使用的是你电脑上运行的操作系统的说明（MAC、Windows或Linux）。

### 3. 更新你的远程以支持SSH

如果你一直使用用户名/密码登录来验证GitHub，而现在改用基于SSH的认证，你可能需要更新连接到GitHub repo的远程。一个快速了解情况的方法是

1. 在命令行中，将光盘放到 repo 目录中（cd path/to/practice-git-skillz）。
2. 当进入感兴趣的目录时，运行`git remote -v`

如果返回的URL看起来像下面这个，以**https://** 开始，那么你仍然在使用密码或令牌进行认证。如果你设置了基于令牌的认证，那么你可以使用https:// 远程设置。



然而，如果你希望使用SSH，那么你将需要为你的 repo 更新远程。



要把远程仓库的网址从 https 改为 SSH，请执行以下操作

1. 转到GitHub.com上的仓库
2. 使用绿色的代码按钮，复制该仓库的SSH URL
3. 使用 `git remote set-url` 在本地更新该网址

***
$ git remote set-url origin git@github.com:your-username/practice-git-skillz.git
***

一旦你更新了你的远程，SSH认证就应该工作了。

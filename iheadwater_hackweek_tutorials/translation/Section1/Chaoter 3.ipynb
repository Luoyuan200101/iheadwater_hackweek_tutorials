### Ⅰ. 用于开放可重复科学的 Jupyter Notebook

如果您还记得关于开放可重复科学的第一章，Jupyter 是一个开源平台，其中包含一套工具，包括：

- Jupyter Notebook：一个基于浏览器的应用程序，允许您创建和共享包含实时代码、方程式、可视化和叙述性文本的文档（即 Jupyter Notebook 文件）。
- Jupyter Lab：一个基于浏览器的应用程序，允许您访问多个 Jupyter Notebook 文件以及其他代码和数据文件
- Jupyter Hub：可以在服务器上运行的多人版 Jupyter Notebook and Lab。

Jupyter Notebook 文件格式 ( .ipynb) 允许您将描述性文本、代码块和代码输出组合在一个文件中。当您运行代码时，它会在笔记本文件中生成输出，包括绘图和表格。然后，您可以将笔记本导出到一个.pdf或.html然后可以与任何人共享。

这种格式很强大，你可以：

1. 与可能想要运行它的任何人共享笔记本本身，并
2. 将笔记本转换为可以像报告一样查看的 PDF 或 HTML 格式

你可以使用 Jupyter Notebooks（.ipynb文件）来记录工作流程并共享代码以进行数据处理、分析和可视化。

#### 1. 在 Jupyter Notebooks 中结合 Markdown 和代码来记录工作流程

在 Jupyter Notebooks 中使用 Markdown 和 Code 有很多优点：

- **可读性**：阅读网页或包含文本和数字的报告要容易得多。通过在代码周围添加 Markdown 文本，你的项目变得更加友好且更易于理解。
- **简单语法**： Markdown 使用可以快速学习的简单格式化语法。这减少了设计有据可查的 Jupyter Notebooks 的学习曲线。
- **对未来自我的有用提醒**：在编写代码时，请考虑未来的自我。如果你为未来的自己留下一组记录良好的 Jupyter Notebook 文件，这些文件既运行您的代码又描述了这些步骤，那么你工作的所有组件都会清楚地记录在案。你（和你未来的自己）不必记住完成工作流程时使用了哪些步骤、假设或测试。
- **易于修改**：你可以通过修改现有代码块或添加新代码块轻松扩展或优化 Jupyter Notebook 文件中包含的分析。
- **灵活的导出格式**：存储在 Jupyter Notebook 文件中的分析结果可以多种格式传播，包括 HTML、PDF 和幻灯片。
- **易于共享**：如果你的所有分析都包含在一个或多个笔记本中并进行了描述，则可以轻松与人共享。你的同事也可以轻松复制你的工作流程。

#### 2. Jupyter Notebooks 对你的同事有益

数据、代码和结果之间的联系使 Jupyter Notebook 功能强大。你可以与同事共享整个工作流程，他们可以快速查看你的流程。

你还可以使用包含代码和数据分析结果的 .ipynb 文件编写报告。为了丰富文档，你可以添加文本，就像在 Word 文档中一样，描述你的工作流程、讨论结果并呈现结论 - 以及分析结果。

#### 3. Jupyter Notebooks 对你和你未来的自己都有好处

Jupyter notebooks ( .ipynb) 是高效的。如果你需要对工作流程进行更改，你只需修改代码并再次运行报告即可。

Jupyter Notebooks 允许你添加文档以提醒自己你的过程。此外，你使用的所有代码都在笔记本中，随时可以重新运行或修改。

#### 4. 将笔记本转换为可共享的 HTML 文件

你可以将包含 Code 和 Markdown 的 Jupyter Notebooks 保存为.html文件。当你将笔记本另存为.html时，它会创建一个呈现精美的网页，其中包含可见的代码和输出以及呈现格式精美的 Markdown 文本。这意味着你可以与同事共享文件，他们可以看到你的整个工作流程 - 无需重新运行你的分析。你将在本教科书后面学习如何导出到.html其他格式。

### Ⅱ.什么是 Jupyter Notebook？

Jupyter Notebook是一个集成开发环境 (IDE)，允许编写代码、在计算机上导航文件、检查变量等。

IDE 不同于文本编辑器，因为 IDE 允许你以交互方式编写、测试和调试代码。

虽然还有其他常用的PythonIDE，例如 Spyder（随PythonAnaconda 发行版一起提供）和 PyCharm，但在本教科书中，你将使用Jupyter Notebook它作为你的 IDE。

#### 1. Jupyter Notebook的优势

Jupyter Notebook为用户提供有用的功能，包括：

- 以交互方式编写和运行代码。
- 使用良好的编程代码和文本的组合记录工作流的所有方面。
- 与可以查看所有代码、代码输出和文本说明的同事共享工作。

此功能通过促进工作流程上的文档和协作来支持开放的可重复科学。

#### 2. Jupyter Notebook 让编码和文档更高效

Jupyter Notebook可以帮助节省时间：

- **快捷方式**：Jupyter Notebook有不同的快捷方式用于运行单元格、添加新单元格和格式化笔记本，节省时间。
- **自动完成**：自动完成可用于快速查找函数名称、这些函数的参数以及存储在你的环境中的变量。这使得输入代码更容易，更不容易出错。在打字时按下Tab按钮，会提示Jupyter Notebook帮助你完成你要调用的变量或函数的名称。
- **Markdown 集成**：你可以将文本和代码组合在同一个文档中。这使可以记录和解释你编写的代码，并使工作流程更具可重复性。

同时 Jupyter Notebook 也是一个开源产品

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create data
boulder_precip = pd.DataFrame(columns=["date", "precip"], 
                              data=[
                                  ["2013-09-09", 0.1], ["2013-09-10", 1.0], 
                                  ["2013-09-11", 2.3], ["2013-09-12", 9.8], ["2013-09-13", 1.9],
                                  ["2013-09-14", 0.01], ["2013-09-15", 1.4], ["2013-09-16", 0.4]])      
# Create plot
fig, ax = plt.subplots()
ax.bar(boulder_precip['date'].values, boulder_precip['precip'].values)
ax.set(title="Daily Precipitation (inches)\nBoulder, Colorado 2013", 
       xlabel="Date", ylabel="Precipitation (Inches)")
plt.setp(ax.get_xticklabels(), rotation=45);
```

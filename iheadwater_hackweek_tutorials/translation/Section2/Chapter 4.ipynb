### Ⅰ. 什么是Markdown

Markdown是一种人类可读的语法（也被称为标记语言），用于格式化文本文件。Markdown可以用来制作格式化良好的文档，包括PDF和网页。

当你在文档中使用Markdown格式化文本时，它类似于使用像Microsoft Word或Google Docs这样的文字处理工具中的格式化工具（如粗体、标题1、标题2）。然而，你不是使用按钮来应用格式化，而是使用语法，如** 这个语法在markdown中对文本加粗 ** 或 # 这里是一个标题。

Markdown语法允许你以多种方式格式化文本，如制作标题、加粗和斜体字、创建列表、添加链接、格式化数学符号和制作表格。这些选项使你能够以视觉上吸引人的和有组织的方式格式化文本，以展示你的想法。

你可以在许多不同的工具中使用Markdown来格式化文本，包括[GitHub.com](https://github.com/)，使用RMarkdown的R，以及Jupyter Notebook，你将在这个页面上了解更多。

#### 1. Jupyter Notebook 中的Markdown

Jupyter Notebook 的一大好处是，它允许你在一个文件中结合代码（如Python）和Markdown，这样你就可以轻松地记录你的工作流程。

一个Jupyter Notebook 文件使用单元格来组织内容，它既可以包含渲染使用Markdown语法编写的文本的单元格，也可以包含和运行Python代码的单元格。

### Ⅵ. 清理表格式文本文件，以便在Python中打开它们

#### 1. 文本文件中缺失的数据值和标题

不是所有的文本文件都像上面的例子那样简单。许多文本文件在数据上方有几行标题文本，为你提供关于数据本身的有用信息。这些数据被称为元数据。

另外，很多时候，在收集的数据中会有数据缺失。这些缺失的数值将使用一个特定的数值来识别，希望在该文件的元数据中有所记载。

接下来你将探讨一些需要清理的温度数据。

```python
miami_temp_url = "https://www.ncdc.noaa.gov/cag/city/time-series/USW00012839-tmax-12-12-1895-2020.csv"

miami_temp = pd.read_csv(miami_temp_url)
miami_temp
```

75行×4列

请注意，上面的数据包含一些额外的行信息。然而，这些信息对你的理解很重要。

- 缺少:-99 ———— 这是代表 "无数据 "的数值。如果一个传感器停止工作或一个测量没有被记录，可能会出现错误的数据。你将想删除任何缺失的数据值。
- 单位:华氏度--在你试图解释数据所显示的内容之前，首先了解数据的单位总是很重要的!

下面你将使用存储在标题中的所有信息来导入你的数据。你还将删除前几行的数据，因为它们实际上不包含任何数据值。这些行包含元数据。

#### 2. Python中的函数参数

参数指的是你在Python中运行一个函数时可以指定的选项。你可以用上面调整绘图类型和颜色的方法来调整与导入数据相关的参数。

下面你使用:

- skiprows=: 告诉Python跳过数据的前3行。
- na_values=: 告诉Python将任何缺失的数据值重新分配为 "NA"

NA指的是缺失的数据。当你把一个值指定为NA (NaN或Python中的Not a Number)时，它将不包括在绘图或任何数学运算中。

```python
# Open the Miami data skipping the first 3 rows and setting no data values
miami_temp = pd.read_csv(miami_temp_url,
                         skiprows=3,
                         na_values=-99)

# View the first 5 rows of the data
miami_temp.head()
```

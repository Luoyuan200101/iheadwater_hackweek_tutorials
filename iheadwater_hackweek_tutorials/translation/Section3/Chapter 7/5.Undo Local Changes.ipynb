## Ⅰ. 撤消更改（在 git add 之前）

如果你更改了文件但尚未运行`git add`，你可以通过运行来撤消更改`git checkout`。你可以使用以下命令指定特定文件：

`git checkout filename.ipynb`

或者你可以撤消所有更改（包括所有尚未提交的更改！）：

`git checkout .`

## Ⅱ. 取消暂存更改（在 git add 之后，在 git commit 之前）

请记住，一旦使用向版本控制添加一组更改`git add`，更改的文件就已暂存。如果已通过暂存更改的文件`git add`，则需要使用`git reset`拉取文件的最新提交版本并撤消所做的更改。

按照以下步骤练习撤消在`git add`之后但之前发生的更改`git commit`。首先，对 README.md 进行新的更改，然后运行以暂存`git add`更改的文件（即添加到版本控制）。检查`git status`以查看更改的状态。

请参阅下面的示例来更改文件使用`bash`（您不需要在 # 之后键入注释）：

***
#Comment in bash; modify the README file
$ echo 'Some more changes' >> README.md
***

然后，运行`git add`暂存文件（即添加到版本控制）并`git status`再次运行:

***
$ git add README.md

$ git status
***



幸运的是，`git status`再次的输出为我们提供了如何撤消分阶段更改的提示。你可以使用`git reset HEAD filename`取消暂存更改。HEAD指最近提交的文件版本：

***
$ git reset HEAD README.md
***

返回：



当你使用`git reset`时，你的更改仍然存在于文件中，但文件已被取消暂存（即版本控制不再跟踪更改）。所以现在，`git status`将显示一条消息，表明有可以添加或暂存到版本控制的更改。因此，就像你根本没有跑过一样`git add`。

***
$ git status
***



现在你有了未暂存的更改（即未在版本控制中跟踪），你可以`git checkout`再次使用来撤消这些更改。

***
$ git checkout README.md

$ git status
***

返回你的存储库中没有更改：



`Git reset`本质上是命令的反面`git add`。它将撤消`git add`从版本控制中删除更改的文件，然后您可以`git checkout`撤消文件中的更改。

## Ⅲ. 撤消提交（在 git commit 之后，在 git push 之前）

如果你已经提交了对文件的更改（即您同时运行了`git add`和`git commit`），并且想要撤消这些更改，那么你可以使用`git reset HEAD~`撤消你的提交。

与上一个示例类似，当你使用`git reset HEAD~`时，修改将被取消暂存，然后你可以使用`git checkout`撤消对文件的更改。

要练习撤消提交，请对 README.md 进行另一次更改。

然后，运行`git add`以暂存文件（即添加到版本控制），然后运行`git commit`以提交文件。检查`git status`以查看更改的状态。

***
#Comment in bash; modify the README file
$ echo 'Even more changes.' >> README.md

$ git commit -m 'Update description in README.md'

$ git status
***

请注意，现在你的文件已取消暂存，因此不再被跟踪！

此时，你可以根据需要运行`git checkout README.md`以撤消对文件的更改。

***
$ git checkout README.md

$ git status
***
